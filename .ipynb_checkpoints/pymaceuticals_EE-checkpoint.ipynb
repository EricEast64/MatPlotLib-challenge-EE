## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_data = pd.merge(mouse_metadata, study_results, on ="Mouse ID", how="outer")

# Display the data table for preview
merged_data.head()
```

```python
# Checking the number of mice.
mice_count = merged_data["Mouse ID"].value_counts()
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_data = merged_data[merged_data.duplicated(["Mouse ID", "Timepoint"])]

duplicate_count = duplicate_data["Mouse ID"].value_counts()
duplicate_count
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = merged_data.drop_duplicates(subset = ["Mouse ID", "Timepoint"])
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count_clean = clean_data["Mouse ID"].value_counts()
mice_count_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
group_data = clean_data.groupby(["Drug Regimen"])

group_mean = group_data["Tumor Volume (mm3)"].mean()
group_median = group_data["Tumor Volume (mm3)"].median()
group_variance = group_data["Tumor Volume (mm3)"].var()
group_std = group_data["Tumor Volume (mm3)"].std()
group_sem = group_data["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.
drug_data = pd.DataFrame()

drug_data["Mean"] = group_mean
drug_data["Median"] = group_median
drug_data["Variance"] = group_variance
drug_data["Standard Deviation"] = group_std
drug_data["SEM"] = group_sem

drug_data
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

tumor_data = clean_data.drop(clean_data.columns[[0,1,2,3,4,5,7]], axis = 1)

aggregate_data = tumor_data.aggregate(['mean', 'median', 'var', 'std', 'sem'])
aggregate_data = aggregate_data.transpose()

aggregate_data.head()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(Capomulin_df["Tumor Volume (mm3)"],Capomulin_df["Weight (g)"])
print(f"The correlation between weight and tumor volume is {round(correlation[0],2)}")
```

```python
m_slope, m_int, m_r, m_p, m_std_err = st.linregress(Capomulin_df["Tumor Volume (mm3)"], Capomulin_df["Weight (g)"])
m_fit = m_slope * Capomulin_df["Tumor Volume (mm3)"] + m_int

plt.scatter(Capomulin_df["Tumor Volume (mm3)"], Capomulin_df["Weight (g)"],)
plt.plot(Capomulin_df["Tumor Volume (mm3)"], m_fit)
plt.xlabel("Tumor Volume (mm3)")
plt.ylabel("Mouse Weight (g)")
```
